```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/weather_hist_iowa.csv")
data
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns=['snow', 'wpgt', 'tsun', 'coco'])
data
```

```python
# Check and handle missing values for all types
for col in data.columns:
    if data[col].dtype == 'float64':
        data[col].fillna(data[col].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates()
```

```python

plt.figure()
sns.heatmap(data[['temp', 'dwpt', 'rhum', 'prcp', 'wdir', 'wspd', 'pres']].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
elec_data = pd.read_csv("data/MW_electricity_cleaned.csv")
elec_data
```

```python
data = data.rename(columns={'time': 'UTC Time at End of Hour'})  

df_merged = pd.merge(elec_data, data, on='UTC Time at End of Hour', how='left')

df_merged
```

```python
# Define the columns for each DataFrame
columns_non_renewables = ['UTC Time at End of Hour', 'Net Generation (MW) from Coal', 'Net Generation (MW) from Natural Gas', 'Net Generation (MW) from Nuclear', 'Net Generation (MW) from All Petroleum Products', 'Net Generation (MW) from Other']
columns_renewables = ['UTC Time at End of Hour', 'Net Generation (MW) from Hydropower and Pumped Storage', 'Net Generation (MW) from Solar', 'Net Generation (MW) from Wind', 'temp', 'dwpt', 'rhum', 'prcp', 'wdir', 'wspd', 'pres']

non_renewables = df_merged[columns_non_renewables]
renewables = df_merged[columns_renewables]
```

```python
non_renewables
```

```python
renewables
```

```python
renewables.iloc[:,1:].rename(columns={'Net Generation (MW) from Hydropower and Pumped Storage':'Net Generation (MW) from Hydro'}).corr()
```

```python
plt.figure()
sns.heatmap(renewables.iloc[:,1:].rename(columns={'Net Generation (MW) from Hydropower and Pumped Storage':'Net Generation (MW) from Hydro'}).corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
filepath = 'data/non_renewables_cleaned.csv'
filepath2 = 'data/renewables_cleaned.csv'

non_renewables.to_csv(filepath, index=False)
renewables.to_csv(filepath2, index=False)
```
